```python
def fetch_data(data_url):
    # URL to fetch data
    # data_url = "https://sdmx.oecd.org/public/rest/data/OECD.ECO.MAD,DSD_EO@DF_EO,/USA.CPI_YTYPCT+GDPV_ANNPCT.Q?startPeriod=1990-Q1&dimensionAtObservation=AllDimensions

    # Send a GET request to the API
    response = requests.get(data_url)

    # Check the status code
    if response.status_code == 200:
        # Print the response content type
        content_type = response.headers['Content-Type']
        print(f"Content Type: {content_type}")

        # Check if the response is XML
        if 'xml' in content_type:
            print("Response is in XML format.")

            # Parse the XML content
            root = ET.fromstring(response.content)

            # Extract data
            data = []
            namespace = {'ns0': 'http://www.sdmx.org/resources/sdmxml/schemas/v2_1/message',
                         'ns2': 'http://www.sdmx.org/resources/sdmxml/schemas/v2_1/data/generic'}

            for obs in root.findall('.//ns2:Obs', namespace):
                time_period = obs.find('ns2:ObsKey/ns2:Value[@id="TIME_PERIOD"]', namespace).get('value')
                measure = obs.find('ns2:ObsKey/ns2:Value[@id="MEASURE"]', namespace).get('value')
                value = obs.find('ns2:ObsValue', namespace).get('value')
                data.append((time_period, measure, float(value)))

            # Create a DataFrame
            df = pd.DataFrame(data, columns=['Time Period', 'Measure', 'Value'])

            # Sort by Time Period and group by Measure
            df.sort_values(by='Time Period', inplace=True)
#             grouped = df.groupby('Measure')

#             # Write to CSV file
#             with open('sorted_grouped_data.csv', 'w') as file:
#                 for measure, group in grouped:
#                     file.write(f'Measure: {measure}\n')
#                     group.to_csv(file, index=False, header=True)
#                     file.write('\n')

#             print("Data has been written to sorted_grouped_data.csv")

            return df

        else:
            print("Unhandled content type.")
    else:
        print(f"Failed to retrieve the data. Status code: {response.status_code}")
```

```python
import re
from datetime import datetime

def extract_quarter(date_string):
    # Dictionary to map months to their respective quarters
    month_to_quarter = {
        1: "Q1", 2: "Q1", 3: "Q1",
        4: "Q2", 5: "Q2", 6: "Q2",
        7: "Q3", 8: "Q3", 9: "Q3",
        10: "Q4", 11: "Q4", 12: "Q4"
    }

    # Extract month and year from the string using regex
    match = re.search(r'(\b\w+\b)\s+(\d{4})', date_string)
    if match:
        month_str = match.group(1)
        year = match.group(2)

        # Convert month name to month number
        try:
            month_num = datetime.strptime(month_str, '%B').month
        except ValueError:
            raise ValueError(f"Invalid month name found in the string: {month_str}")

        # Determine the quarter
        quarter = month_to_quarter[month_num]
        
        # Return the formatted quarter-year string
        return f"{quarter}-{year}"
    else:
        raise ValueError("Date string does not contain a valid month and year.")
```

```python
def find_country_nums(data, country_codes: list, verbose=False):
    if (type(country_codes)!= tuple) and (type(country_codes)!=list):
        raise ValueError('country codes must be a list or tuple')
    country_data = data['structure']['dimensions']['series'][0]['values']
    country_lst = []
    for code in country_codes:
        country_num = None
        for i in range(len(country_data)):
            if country_data[i]['id'] == code:
                if verbose:
                    print(country_data[i], ":", i)
                country_num = (code, i)
                    
        if country_num:
            country_lst.append(country_num)
        else: print(f"Could not find country with code {code}")
        
    return country_lst
                
def find_variable_nums(data, variable_codes: list, verbose=False):
    if (type(variable_codes)!= tuple) and (type(variable_codes)!=list):
        raise ValueError('variable codes must be a list or tuple')
    variable_data = data['structure']['dimensions']['series'][1]['values']
    var_lst = []
    for code in variable_codes:
        var_num=None
        for i in range(len(variable_data)):
            if variable_data[i]['id'] == code:
                if verbose:
                    print(variable_data[i], ":", i)          
                var_num = (code, i)
                    
        if var_num:
            var_lst.append(var_num)
        else: print(f"Could not find variable with code {code}")
                    
                
    return var_lst
            
def find_freq_nums(data, freq_codes: list, verbose=False):
    if (type(freq_codes)!= tuple) and (type(freq_codes)!=list):
        raise ValueError('Frequency codes must be a list or tuple')
    freq_data = data['structure']['dimensions']['series'][2]['values']
    freq_lst = []
    for code in freq_codes:
        freq_num = None
        for i in range(len(freq_data)):
            if freq_data[i]['id'] == code:
                if verbose:
                    print(freq_data[i], ":", i)
                freq_num = (code, i)
                    
        if freq_num:
            freq_lst.append(freq_num)
        else: print(f"Could not find frequency with code {code}")
                
                
    return freq_lst
            
def find_time_series(data, country_codes, var_codes, freq_codes, country_verbose=False, var_verbose=False, freq_verbose=False, time_series_verbose=False):
    country_lst = find_country_nums(data, country_codes, verbose=country_verbose)
    var_lst = find_variable_nums(data, var_codes, verbose=var_verbose)
    freq_lst = find_freq_nums(data, freq_codes, verbose=freq_verbose)
        
    df = pd.DataFrame
    
    if time_series_verbose:
        print("country_lst", country_lst)
        print("var_lst", var_lst)
        print("freq_lst", freq_lst)
        
    # info = data['structure']['name'].strip(".").split()[-2:]
    # if info[0] in ['April', 'May', 'June']:
    #     publish_date = f'Q2-{date[1]}'
    # elif info[0] in ['October', 'November', 'December']:
    #     publish_date = f'Q4-{date[1]}'
    
    publish_date = extract_quarter(data['structure']['name'])
        
    for country in country_lst:
        for var in var_lst:
            for freq in freq_lst:
                code = f"{country[1]}:{var[1]}:{freq[1]}"
                # print(code)
                try:
                    key_lst = list(data['dataSets'][0]['series'][code]['observations'].keys())
                except Exception as e:
                    print(f"There was an error for country code {country}, variable {var}, frequency {freq}")
                    print(repr(e))
                    break
                
                    
#                 start_key = int(key_lst[0])
#                 end_key = int(key_lst[-1])
               
                date_code_lst = data['structure']['dimensions']['observation'][0]['values'] 
                
#                 start_date = date_code_lst[start_key]
#                 end_date = date_code_lst[end_key]
                
#                 print(start_date, end_date)
                
                dictionary = data['dataSets'][0]['series'][code]['observations']
                
                temp_df = pd.DataFrame(dictionary).transpose().drop(1, axis=1)
                temp_df.rename(columns={0: var[0]}, inplace=True)
                temp_df = pd.concat([temp_df], axis=1, keys=[country[0]])
                temp_df = pd.concat([temp_df], axis=1, keys=[publish_date])

                if df.empty:
                    df = temp_df
                else: df = pd.concat([df, temp_df], axis=1)
                
    if not df.empty:            
        df.index = (df.index.map(lambda x:date_code_lst[int(x)]['name']))

                
    return df
    

# find_time_series(data, ['USA'], ['GDP_ANNPCT', 'CPI_YTYPCT'], ['Q'])
```

```python
def get_imf_weo_data(report_year, report_month, country_codes: list, subject_codes: list, start_year = 1980, predict_years=0, verbose=False):
    month_to_quarter = {
        'January': 'Q1', 'February': 'Q1', 'March': 'Q1',
        'April': 'Q2', 'May': 'Q2', 'June': 'Q2',
        'July': 'Q3', 'August': 'Q3', 'September': 'Q3',
        'October': 'Q4', 'November': 'Q4', 'December': 'Q4'
    }
    import requests
    import pandas as pd
    base = 'https://www.imf.org/en/Publications/WEO/weo-database/'
    start_year = '&sy='+ str(start_year)
    end_year = '&ey='+str(report_year)
    
    report_quarter = month_to_quarter[report_month]
    report_date = f'{report_quarter}-{report_year}'

    if predict_years:
        if predict_years<=5:
            end_year = '&ey='
            end_year += str(report_year + predict_years)
        else:
            print('predict_years must be <=5')
    report_year = str(report_year)+"/"
    report_month = str(report_month)+"/"
    
    country = 'weo-report?c='
    for code in country_codes:
        country+=(str(code)+",")
        
    subject = '&s='
    for code in subject_codes:
        subject+=(str(code)+",")
        
    url = f"{base}{report_year}{report_month}{country}{subject}{start_year}{end_year}&ssm=1&scsm=1&scc=1&ssd=1&ssc=1&sic=1&sort=country&ds=.&br=1"
    if verbose: print(url)
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to load page {url}")

    # Load the data into a pandas DataFrame
    try:
        data = pd.read_html(response.content)
    except Exception as e:
        print(f"There was an error read the page content")
        print(url)
        print(repr(e))
        return pd.DataFrame()
        

    # Since the page may contain multiple tables, select the appropriate one
    # Here, we assume the first table contains the data we need
    df = data[0]
    # drop excess columns
    df.drop(['Scale', 'Subject Notes', 'Country/Series-specific Notes'], axis=1)
    
    # create multi level column indexes
    
    col_names = df.copy().loc[:, :'1980'].T.loc[['ISO', 'WEO Subject Code']].values
    
    # Step 3: Create a MultiIndex from dummy1 values and add report_date
    columns = pd.MultiIndex.from_arrays(
        [[report_date] * len(col_names[0]), col_names[0], col_names[1]],
        names=['Report Date', 'Country', 'Indicator']
    )

    # Step 4: Transpose dummy2 and assign the multi-level column headers
    df = df.loc[:, '1980':].T
    df.columns = columns
    
    return df
```

```python
df_imf=pd.DataFrame()
for year in range(2024, 1990, -1):
    for month in ['October', 'September', 'May', 'April']:
        df_new = get_imf_weo_data(year, month, ['111', '924'], ['NGDP_RPCH','PCPIPCH'], predict_years=5)
        if not df_new.empty:
            df_imf = pd.concat([df_imf, df_new], axis=1)
df_imf.to_csv('imf_projection.csv')
```
